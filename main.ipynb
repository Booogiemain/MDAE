# "Pré-processamento de Dados para Mineração: Uma Abordagem Prática com a Base de Dados: Interrupções de Energia Elétrica nas Redes de Distribuição"

<h2>Introdução</h2>

Na mineração de dados, o pré-processamento é o ponto de partida essencial para garantir a qualidade e a utilidade dos dados que serão analisados. É o processo de limpeza, organização e preparação dos dados brutos para análise posterior. Sem um pré-processamento adequado, os dados podem conter ruído, inconsistências e valores ausentes, o que pode levar a conclusões errôneas e modelos imprecisos. Portanto, entender a importância do pré-processamento de dados é fundamental para garantir resultados confiáveis e significativos na mineração de dados. Nesta introdução, exploraremos brevemente o papel crucial do pré-processamento na extração de informações valiosas a partir de conjuntos de dados.

<h2>Objetivo</h2>

O objetivo deste trabalho é aplicar técnicas de pré-processamento em uma base de dados específica, com foco na preparação dos dados para análises futuras relacionadas ao tema "Interrupções de Energia Elétrica nas Redes de Distribuição". Essas técnicas visam limpar, organizar e transformar os dados brutos em um formato mais adequado e compreensível, de modo a facilitar a extração de informações relevantes sobre interrupções de energia elétrica. Ao realizar o pré-processamento, buscamos identificar e corrigir inconsistências, preencher valores ausentes, remover outliers e transformar os dados conforme necessário para facilitar análises estatísticas, modelagem preditiva ou qualquer outra forma de análise exploratória. O objetivo final é preparar a base de dados de forma que possa fornecer insights significativos e precisos sobre os padrões, causas e impactos das interrupções de energia elétrica nas redes de distribuição, contribuindo assim para melhorias na gestão e manutenção dessas redes.

<h3>Pré-processamento dos dados no código fornecido:</h3>

**Seleção de Colunas Relevantes:**
No código, são especificadas as colunas necessárias para a análise através da lista columns_needed. Isso ajuda a reduzir a quantidade de dados a serem carregados e processados, tornando o processo mais eficiente.

**Especificação dos Tipos de Dados:**
O código especifica os tipos de dados adequados para cada coluna utilizando o parâmetro dtype ao ler o arquivo CSV. No entanto, todos os tipos de dados estão definidos como str, o que pode ser melhorado, especialmente para as colunas de datas.

**Leitura do Arquivo CSV:**
O arquivo CSV é lido utilizando a função pd.read_csv(). Durante a leitura, são aplicadas as configurações de encoding, separador, seleção de colunas e tipos de dados especificados.

**Remoção de Valores em Branco:**
O código remove todas as linhas com valores em branco na coluna 'NumOrdemInterrupcao' utilizando o método dropna(). Isso pode ser útil para garantir que apenas dados completos sejam utilizados na análise.

**Agrupamento e Análise dos Dados:**
Os dados são agrupados pela coluna 'NomAgenteRegulado' e é contado o número de valores distintos na coluna 'NumOrdemInterrupcao'. Essa análise fornece informações sobre a distribuição dos valores e permite identificar os agentes com o maior número de ocorrências.

**Visualização dos Resultados:**
Os resultados da análise são visualizados através de um gráfico de barras, onde são mostrados os cinco primeiros 'NomAgenteRegulado' com o maior número de 'NumOrdemInterrupcao'. Isso facilita a compreensão dos resultados e permite uma rápida identificação dos principais agentes.

<h2>Desenvolvimento</h2>

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime


# Carregar o arquivo CSV
file_path = 'C:/Users/u11886/Desktop/MINERAÇÃO DE DADOS P1/interrupcoes-energia-eletrica-2024-datasetdefault.csv'
columns_needed = ['NumOrdemInterrupcao', 'NomAgenteRegulado', 'DscFatoGeradorInterrupcao', 'DatFimInterrupcao', 'DatInicioInterrupcao', 'DscSubestacaoDistribuicao', 'DscAlimentadorSubestacao']
dtype = {'NumOrdemInterrupcao': str, 'NomAgenteRegulado': str, 'DscFatoGeradorInterrupcao': str, 'DatFimInterrupcao': str, 'DatInicioInterrupcao': str, 'DscSubestacaoDistribuicao': str, 'DscAlimentadorSubestacao': str}
df_init = pd.read_csv(file_path, encoding='ISO-8859-1', sep=';', low_memory=False, usecols=columns_needed, dtype=dtype)

# Remove todas as linhas com valores em branco na coluna 'NumOrdemInterrupcao'
df = df_init.dropna(subset=['NumOrdemInterrupcao'])

# Agrupar os dados pela coluna 'NomAgenteRegulado' e contar os valores distintos na coluna 'NumOrdemInterrupcao'
grouped = df.groupby('NomAgenteRegulado')['NumOrdemInterrupcao'].nunique()

# Ordenar o DataFrame agrupado em ordem decrescente
grouped_sorted = grouped.sort_values(ascending=False)

# Pegar as 10 primeiras 'NomAgenteRegulado'
top_5 = grouped_sorted.head(10)

# Criar o gráfico
plt.figure(figsize=(10,6))
bars = top_5.plot(kind='bar')

plt.title('Top 10: Interrupções por distribuidora no Brasil no primeiro trimestre de 2024')
plt.xlabel('Distribuidora')
plt.ylabel('Contagem Distinta de Interrupções')

# Adicionar rótulos de dados
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 2), ha='center', va='bottom')

plt.show()
```

<h3>Comentários Figura 1</h3>

As distribuidoras que mais possuem ocorrências no primeiro trimestre do ano de 2024 são CEMIG, COPEL, LIGHT, EQUATORIAL GOIÁS E AMPLA ENERGIA. A que lidera a posição negativa em relação ao cenário nacional é a CEMIG com 177.858 mil ocorrências no primeiro trimestre do ano vigente.

```python
# Agrupar os dados pela coluna 'NomAgenteRegulado' e contar os valores distintos na coluna 'NumOrdemInterrupcao'
grouped = df.groupby('DscFatoGeradorInterrupcao')['NumOrdemInterrupcao'].nunique()

# Ordenar o DataFrame agrupado em ordem decrescente
grouped_sorted = grouped.sort_values(ascending=False)

# Pegar as 5 primeiras 'NomAgenteRegulado'
top_5 = grouped_sorted.head(5)

# Criar o gráfico
plt.figure(figsize=(10,6))
bars = top_5.plot(kind='bar')

plt.title('Top 5: Fator gerador de interrupções no Brasil no primeiro trimestre de 2024')
plt.xlabel('Fator gerador')
plt.ylabel('Contagem Distinta de interrupções')

# Adicionar rótulos de dados
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, round(yval, 2), ha='center', va='bottom')

plt.show()
```

<h3>Comentários Figura 2</h3>

O fator gerador que mais impactou as ocorrências no primeiro trimestre foram por causas de falhas de materiais ou equipamentos e por causas climáticas de ventos, sendo assim, as ocorrências desse genêro tem natureza plausível de manutenção preventiva e corretiva.

```python
%matplotlib notebook
import matplotlib.pyplot as plt

df = df.copy()


df['DatFimInterrupcao'] = pd.to_datetime(df['DatFimInterrupcao'], format="%Y-%m-%d %H:%M:%S")
df['DatInicioInterrupcao'] = pd.to_datetime(df['DatInicioInterrupcao'], format="%Y-%m-%d %H:%M:%S")

# Calcula a diferença de tempo
df['DiferencaTempo'] = df['DatFimInterrupcao'] - df['DatInicioInterrupcao']

# Converte a diferença de tempo para horas
df['DiferencaTempo'] = df['DiferencaTempo'].dt.total_seconds()/3600

# Calcula o tempo médio para cada 'NumOrdemInterrupcao'
media_tempo = df.groupby('NumOrdemInterrupcao')['DiferencaTempo'].mean()

# Seleciona os três números de ordem de interrupção com as maiores médias de tempo de interrupção
top_5_numeros_ordem = media_tempo.nlargest(3)

# Cria uma figura e um eixo
fig, ax = plt.subplots()

# Plota os dados
top_5_numeros_ordem.plot(kind='line', ax=ax)

# Define os rótulos dos eixos
ax.set_xlabel('Interrupções')
ax.set_ylabel('Média do Tempo de Interrupção (horas)')

# Otimiza o espaçamento
plt.tight_layout()

# Mostra o gráfico
plt.show()
```

<h3>Comentários Figura 3</h3>

O gráfico acima demonstra o tempo médio de interrupções em hora por ocorrência, o interesse nesse tipo de visão consiste em localizar as interrupções que mais afetaram no DEC e FEC por conta de sua elevada duração. A visão acima contempla o **tempo médio de interrupções para as 3 principais ocorrências, considerando todas as distribuidoras do Brasil**.

```python
import matplotlib.pyplot as plt

# Agrupa por 'NomAgenteRegulado' e 'NumOrdemInterrupcao'
grupo = df.groupby(['NomAgenteRegulado', 'NumOrdemInterrupcao'])['DiferencaTempo'].mean()

# Obtém a lista de 'NomAgenteRegulado'
agentes = df['NomAgenteRegulado'].unique()

# Calcula a média do tempo de interrupção para cada agente
medias_por_agente = grupo.groupby('NomAgenteRegulado').mean()

# Seleciona os cinco agentes com as maiores médias
top_5_agentes = medias_por_agente.nlargest(5)

# Cria um gráfico para cada agente
for agente in top_5_agentes.index:
    if agente in grupo and not grupo[agente].isnull().all():
        # Cria uma figura e um eixo com tamanho maior
        fig, ax = plt.subplots(figsize=(10, 6))

        # Plota os dados para o agente atual
        grupo[agente].plot(kind='line', ax=ax)

        # Define os rótulos dos eixos
        ax.set_xlabel('Interrupções')
        ax.set_ylabel('Média do Tempo de Interrupção (horas)')

        # Define o título do gráfico
        ax.set_title(agente)

        # Rotaciona os rótulos do eixo X
        plt.xticks(rotation=45)

        # Otimiza o espaçamento
        plt.tight_layout()

        # Mostra o gráfico
        plt.show()
```

<h3>Comentários Figura 4</h3>

A figura acima contempla o **as 5 principais distribuidoras com maior tempo médio de interrupções do Brasil**. Tendo o mesmo propósito da figura 3, porém expandindo as visões para cada distribuidora de energia, em que cada uma pode localizar as principais ocorrências do estado.

```python
import matplotlib.pyplot as plt

# Extrai o mês da coluna 'DatInicioInterrupcao'
df['Mes'] = df['DatInicioInterrupcao'].dt.month

# Agrupa por 'NomAgenteRegulado' e 'Mes'
grupo = df.groupby(['NomAgenteRegulado', 'Mes'])['DiferencaTempo'].mean().nlargest(6)

# Obtém a lista de 'NomAgenteRegulado' que possuem pelo menos um mês com valor
agentes_com_dados = [agente for agente in df['NomAgenteRegulado'].unique() if agente in grupo and not grupo[agente].isnull().all()]

# Cria um gráfico para cada 'NomAgenteRegulado' que possui dados
for agente in agentes_com_dados:
    # Cria uma figura e um eixo
    fig, ax = plt.subplots()

    # Plota os dados para o 'NomAgenteRegulado' atual
    grupo[agente].plot(kind='bar', ax=ax)

    # Adiciona rótulos de dados a cada barra
    for i, v in enumerate(grupo[agente]):
        ax.text(i, v + 0.01, str(round(v, 2)), ha='center', va='bottom')

    # Define os rótulos dos eixos
    ax.set_xlabel('Mês')
    ax.set_ylabel('Média do Tempo de Interrupção (horas)')

    # Define o título do gráfico
    ax.set_title(agente)
    
    # Rotaciona os rótulos do eixo X
    plt.xticks(rotation=0)

    # Mostra o gráfico
    plt.show()

```

<h3>Comentários Figura 5</h3>

A figura acima contempla o **as 5 principais distribuidoras que tiveram maior tempo médio de interrupções**. Tendo o mesmo propósito da figura 4, porém expandindo as visões para cada mês do primeiro trimestre do ano de 2024.

```python
df = df.copy()

# Concatenar colunas DscSubestacaoDistribuicao e DscAlimentadorSubestacao para criar a coluna SEAL
df['SEAL'] = df['DscSubestacaoDistribuicao'] + df['DscAlimentadorSubestacao']

# Contagem distinta de NumOrdemInterrupcao por SEAL
grouped = df.groupby('SEAL')['NumOrdemInterrupcao'].nunique()

# Ordenar os SEALs pela contagem distinta em ordem decrescente
grouped_sorted = grouped.sort_values(ascending=False)

# Selecionar apenas os 20 primeiros SEALs
top_10_seals = grouped_sorted.head(20)

# Criar o gráfico
plt.figure(figsize=(10, 6))
bars = top_10_seals.plot(kind='bar')

plt.title('Top 20 circuitos com Maior Contagem Distinta de Interrupções')
plt.xlabel('Circuitos')
plt.ylabel('Contagem Distinta de Interrupções')

# Adicionar rótulos de dados
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 0.05, round(yval, 2), ha='center', va='bottom')
    
# Otimiza o espaçamento
plt.tight_layout()

plt.show()
```

<h3>Comentários Figura 6</h3>

A figura acima contempla os principais circuitos(alimentadores) que mais interroperam na rede elétrica em 2024, considerando os circuitos de todas as distribuidoras.
